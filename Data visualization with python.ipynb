```python
import pandas as pd

# Load the dataset
file_path = "C:/Users/LENOVO/Downloads/Africa_climate_change.csv"
df = pd.read_csv(file_path)
```

```python
df
```

```python
# Display basic information about the dataset
print(df.info())

# Check for missing values
missing_values = df.isnull().sum()
print(missing_values)

# Fill missing values or drop them, depending on the data
# For example, let's drop rows with any missing values
df = df.dropna()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter data for Tunisia and Cameroon
tunisia_data = df[df['Country'] == 'Tunisia']
cameroon_data = df[df['Country'] == 'Cameroon']
```

```python
# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(tunisia_data['Year'], tunisia_data['AvgTemperature'], label='Tunisia')
plt.plot(cameroon_data['Year'], cameroon_data['AvgTemperature'], label='Cameroon')
plt.xlabel('Year')
plt.ylabel('Average Temperature')
plt.title('Average Temperature Fluctuations in Tunisia and Cameroon')
plt.legend()
plt.show()
```
